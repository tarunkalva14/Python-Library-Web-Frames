# **Plotting Multiple Lines**

```python
years = range(2014, 2024)
apples = [1.25, 1.50, 1.75, 2.25, 2.75, 3.25, 4.0, 5.25, 5.75, 6.25]
oranges = [6.50, 5.0, 4.75, 4.25, 3.75, 3.50, 2.0, 1.80, 1.70, 0.75]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)');
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')

plt.title("Fruits Yield in India")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, marker='v')
plt.plot(years, oranges, marker='^')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Fruits Yield in India")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=3, ms=8, mew=2, mec='black')
plt.plot(years, oranges, marker='o', c='g', ls=':', lw=2, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Fruits Yiled in India")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o:g')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Fruits Yiled in India")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, 'or')
plt.title("Yield of Apple (tons per hectare)");
```

```python
plt.figure(figsize= (6, 6))
plt.plot(years, apples, 'or')
plt.title("Yield of Apple (tons per hectare)");
```

# **Scatter Plot**

```python
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=flowers_df);
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=flowers_df, hue='species', s=100);
```

```python
plt.figure(figsize=(8,6))
plt.title("Sepals")
sns.scatterplot(x='sepal_length', y='sepal_width', data=flowers_df, hue='species', s=100);
```

```python
plt.plot(flowers_df.sepal_width)
```

```python
flowers_df.sepal_width
```

```python
plt.title("Distrubution of sepal width")
plt.hist(flowers_df.sepal_width);
```

```python
import numpy as np
```

```python
plt.title("Distrubution of sepal width")
plt.hist(flowers_df.sepal_width, bins = np.arange(2, 5, 0.15));
```

```python
plt.title("Distrubution of sepal width")
plt.hist(flowers_df.sepal_width, bins = [1,3,4,6]);
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.5, bins = np.arange(2, 5, 0.25))
plt.hist(virginica_df.sepal_width, alpha=0.5, bins = np.arange(2, 5, 0.25))
```

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
         bins=np.arange(2, 5, 0.25),
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```
